# Image predict

```python
from ultralytics import YOLO
from PIL import Image
import cv2

def resize(img,scale_percent):
#     scale_percent = 30# percent of original size
    width = int(img.shape[1] * scale_percent / 100)
    height = int(img.shape[0] * scale_percent / 100)
    dim = (width, height)
    resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    return resized

img = cv2.imread("data/test/images/IMG_6185.jpg")
model = YOLO('best3.pt')
res = model(img)
res_plotted = res[0].plot()

resized = resize(res_plotted,30)
 
cv2.imshow("result", resized)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Floder Predict

```python
import os
from ultralytics import YOLO
from PIL import Image
import cv2

input_floder = 'data/test/images'  #--------------------
output_floder = 'data/predicted/test'   #--------------------

def resize(img,scale_percent):
#     scale_percent = 30# percent of original size
    width = int(img.shape[1] * scale_percent / 100)
    height = int(img.shape[0] * scale_percent / 100)
    dim = (width, height)
    resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    return resized
```

```python
if not os.path.exists(output_floder):
    os.makedirs(output_floder)
dir_list = os.listdir(input_floder)
```

```python
for i,f in enumerate(dir_list):
    input_path = f'{input_floder}/{f}'
    output_path = f'{output_floder}/{f}'
    
    img = cv2.imread(input_path)
    model = YOLO('best.pt')
    res = model(img)
    res_plotted = res[0].plot()
    # resized = resize(res_plotted,scale_percent)
    cv2.imwrite(output_path, res_plotted)
    
    print(f'{i+1}/{len(dir_list)} : predicted >> {output_path}')
```

# VDO Predict

```python
from ultralytics import YOLO
from PIL import Image
import cv2

cap = cv2.VideoCapture(0)
# cap = cv2.VideoCapture("path/to/your/video/file.mp4")

model = YOLO('best.pt')

while cap.isOpened():
    success, frame = cap.read()

    if success:
        res = model(frame)
        res_plotted = res[0].plot()
        resized = resize(res_plotted,30)
        cv2.imshow("YOLOv8 Inference", res_plotted)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
